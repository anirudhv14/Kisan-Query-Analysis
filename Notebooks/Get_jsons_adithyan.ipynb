Cachar,Assam

```python
http = "http://dackkms.gov.in/Account/API/kKMS_QueryData.aspx?StateCD=02&DistrictCd=0201&Month=6&Year=2020"
```

```python
jsons_yr = {}
months = {1:'Jan', 2:"Feb", 3:"Mar", 4:"Apr", 5:"May", 6:"Jun", 7:"Jul", 8:"Aug", 9:"Sep", 10:"Oct", 11:"Nov", 12:"Dec"}
for i in range(2013,2022):
    jsons = {}
    for j in range(1,13):
        aa = json.loads(requests.get('http://dackkms.gov.in/Account/API/kKMS_QueryData.aspx?StateCD=02&DistrictCd=0201&Month='+str(j)+'&Year='+str(i)).text)
        if(int(aa['ResponseCode']) == 1):
            jsons[months[j]] = aa
    jsons_yr[i] = jsons
```

```python
for i,j in jsons_yr.items():
    fname = "Data\\json\\Cachar_Assam_"
    for k in j.keys():
        print(fname+k+"_"+str(i)+'.json')
        with open(fname+k+"_"+str(i)+'.json', 'w') as outfile:
            json.dump(jsons_yr[i][k], outfile)
```
